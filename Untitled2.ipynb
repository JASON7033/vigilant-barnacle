# output for panda set


```python
import json
from PIL import Image
```

```python
src_annos=json.load(open('anno.json'))
res_annos=json.load(open('coco_PANDA/bbox.json'))
# res_annos=json.load(open('tiny_set_corner_sw640_sh512_test_all_coco/bbox.json'))
result=[]
src_img_id=src_annos['images'][0]['id']
for res_anno in res_annos:
    if res_anno['image_id']==src_img_id:
        result.append(res_anno)
```

```python
import os
import matplotlib.pyplot as plt
src_img_path=os.path.join('image_train',src_annos['images'][0]['file_name'])
img=Image.open(src_img_path)
plt.figure(figsize=(18,9))
color=(0,1,0)
for item in result:
    box=item['bbox']
#     print(box)
    plt.gca().add_patch(plt.Rectangle((box[0]/10, box[1]/10), box[2]/10, box[3]/10, fill=False, color=color))
plt.imshow(img)
plt.show()
```
